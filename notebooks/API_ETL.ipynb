## Discover by Search

```python
# Using the search endpoint to enter keyword and get top results

category = 'general'

search_endpoint_url = f'https://api.tikapi.io/public/search/{category}'

# Here, take the instance of helping a skincare company who wants data on skincare products
params = {
    'query': 'skincare',
    'country': 'us'
}

headers = {
    'X-API-KEY': api_key
}
```

```python
# Sending a get request to a search API endpoint and storing response.
search_api_request = requests.get(search_endpoint_url, params=params, headers=headers)
```

```python
# Empty dictionary of desired columns for video_info table

video_info = {
    'video_id':[],
    'duration':[],
    'collect_count':[],
    'comment_count':[],
    'digg_count':[],
    'play_count':[],
    'share_count':[]
}

# Looping through each results to get the information for each column
for video in search_api_response['data']:
   
    video_id = video['item']['id']
    print(video_id)
    video_info['video_id'].append(video_id)
    
    duration = video['item']['video']['duration']
    print(duration)
    video_info['duration'].append(duration)
    
    collect_count = video['item']['stats']['collectCount']
    print(collect_count)
    video_info['collect_count'].append(collect_count)
    
    comment_count = video['item']['stats']['commentCount']
    print(comment_count)
    video_info['comment_count'].append(comment_count)
    
    digg_count = video['item']['stats']['diggCount']
    print(digg_count)
    video_info['digg_count'].append(digg_count)
    
    play_count = video['item']['stats']['playCount']
    print(play_count)
    video_info['play_count'].append(play_count)
    
    share_count = video['item']['stats']['shareCount']
    print(share_count)
    video_info['share_count'].append(share_count)
    
    print('-'*100)
    
```

```python
# Checking the dictionary
video_info
```

```python
# Converting dictionary into datafram
video_info_df = pd.DataFrame(video_info)
```

```python
# Checking dataframe
video_info_df
```

```python
# Converting data into csv
video_info_df.to_csv('video_info.csv', index=False)
```
