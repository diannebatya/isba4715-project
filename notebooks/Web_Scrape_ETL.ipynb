```python
# importing packages 

# To manipulate data and perform analysis
import pandas as pd 
# To scrape urls 
import requests
# To extract parts
from bs4 import BeautifulSoup
```

```python
# Saving url of webpage into a varibale

url = 'https://www.blume.com/pages/shop'
```

```python
# Getting request to url
request = requests.get(url)
```

```python
# Checking the success of the response
request
```

```python
# Getting the text from the request
request.text
```

```python
# Using beautiful soup to parse the text
soup = BeautifulSoup(request.text, 'html.parser')
```

```python
# For easier readability
print(soup.prettify())
```

```python
# Getting products through the div and class
# collection__item mb15 rel
products = soup.findAll('div', attrs={'class':'collection__item mb15 rel', 'data-module':'collectionItem'})
```

```python
# Checking the first product
products[0]
```

```python
# Getting title
soup.find('h3', class_='medium').text
```

```python
# Getting Price
soup.find('span', class_='money').text
```

```python
# Get rating average and count
rating_span = soup.find('span', class_='junip-product-summary')
rating_average = rating_span['data-product-rating-average']
rating_count = rating_span['data-product-rating-count']
```

```python
# Empty dictionary to append
product_info = {
    'product_id':[],
    'product_name':[],
    'price':[],
    'rating_average':[],
    'rating_count':[]
}
```

```python
# Looping through all products to get desired information

for product in products:
    
    rating_span = product.find('span', class_='junip-product-summary')
    product_id = rating_span['data-product-id']
    print(product_id)
    product_info['product_id'].append(product_id)
    
    product_name = product.find('h3', class_='medium').text
    name = ' '.join(product_name.split())
    print(name)
    product_info['product_name'].append(name)
    
    price = product.find('span', class_='money').text
    print(price)
    product_info['price'].append(price)
    
    
    rating_span = product.find('span', class_='junip-product-summary')
    rating_average = rating_span['data-product-rating-average']
    print(rating_average)
    product_info['rating_average'].append(rating_average)
    
    rating_span = product.find('span', class_='junip-product-summary')
    rating_count = rating_span['data-product-rating-count']
    print(rating_count)
    product_info['rating_count'].append(rating_count)
    
    
```

```python
product_info
```

```python
# Saving result into a dataframe
product_info_df = pd.DataFrame(product_info)
```

```python
# Checking the dataframe
product_info_df
```

```python
# Saving it to a csv 
product_info_df.to_csv('product_info.csv', index=False)
```

## New Dataframe to Support Web Scrape

```python
# To randomize numbers
import random

# Saving product_id list into variable
product_ids = product_info['product_id']

# Generate a list of 50 random product IDs
random_product_ids = [random.choice(product_ids) for product in range(50)]
```

```python
# Generate a 4 digit number and adding it to a set for uniqueness
order_ids = set()
while len(order_ids) < 50:
    order_id = random.randint(1000, 9999)  
    order_ids.add(order_id)

# Convert the set to a list to be used in dataframe
order_ids = list(order_ids)
```

```python
# To format the time
from datetime import datetime, timedelta

# Generate a random date within the year 2023
def generate_random_date():
    start_date = datetime.strptime('2023-01-01', '%Y-%m-%d')
    end_date = datetime.strptime('2023-12-31', '%Y-%m-%d')
    delta = end_date - start_date
    random_days = random.randrange(delta.days + 1)
    random_date = start_date + timedelta(days=random_days)
    return random_date.strftime('%Y-%m-%d')

# Generate a list of 50 random dates
random_dates = [generate_random_date() for date in range(50)]

```

```python
# Generate quantity
random_quantities = [random.randint(1, 5) for quantity in range(50)]
```

```python
# Generate 5 digit customer id and adding to set for uniqueness
customer_ids = set()
while len(customer_ids) < 50:
    customer_id = random.randint(10000, 99999)
    customer_ids.add(customer_id)

# Convert the set to a list to be used in DataFrame
customer_ids = list(customer_ids)
```

```python
# Store results in dictionary
order_info = {
    'order_id': order_ids,
    'product_id': random_product_ids,
    'order_date': random_dates,
    'order_quantity': random_quantities,
    'customer_id': customer_ids
}
```

```python
# Convert into dataframe
order_info_df = pd.DataFrame(order_info)
```

```python
# Checking dataframe
order_info_df 
```

```python
# Converting into csv
order_info_df.to_csv('order_info.csv', index=False)
```
